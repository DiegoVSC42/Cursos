```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013")


axs[0,0].plot(df.loc["A"])
axs[0,0].set_title("A")

axs[0,1].plot(df.loc["B"])
axs[0,1].set_title("B")

axs[1,0].plot(df.loc["C"])
axs[1,0].set_title("C")

axs[1,1].plot(df.loc["D"])
axs[1,1].set_title("D")

for ax in axs.flat:
    ax.set_xlabel("Meses")
    ax.set_ylabel("Número de vendas")

# ymin = 0 
# ymax = 7000

# for ax in axs.ravel():
#     ax.set_ylim(ymin,ymax)
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(14,8))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D")


axs[0,0].plot(df.loc["A"])
axs[0,0].set_title("Vendas na loja A")

axs[0,1].plot(df.loc["B"])
axs[0,1].set_title("Vendas na loja B")

axs[1,0].plot(df.loc["C"])
axs[1,0].set_title("Vendas na loja C")

axs[1,1].plot(df.loc["D"])
axs[1,1].set_title("Vendas na loja D")

for ax in axs.flat:
    ax.set_xlabel("Meses")
    ax.set_ylabel("Número de vendas")

ymin = 0 
ymax = 500

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
plt.show()
```
