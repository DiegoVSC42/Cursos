E mais uma dica: você pode reduzir o tamanho do código utilizando o comando for i, ax in enumerate(axs.flat): que permite um loop iterando sobre todos os subplots da figura. Dentro desse loop você pode passar as funções plot, set_title, set_xlabel, set_ylabel e etc…

```python
import pandas as pd

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criando DataFrame
df = pd.DataFrame(vendas_2022, index=lojas)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2,2,figsize=(14,8))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D",fontsize = 16, fontweight = "bold")


axs[0,0].plot(df.loc["A"],lw=3,color = "red")
axs[0,0].set_title("Vendas na loja A",loc = "left",fontsize = 12,fontweight=600)

axs[0,1].plot(df.loc["B"],lw=3,color = "green")
axs[0,1].set_title("Vendas na loja B",loc = "left",fontsize = 12,fontweight=600)

axs[1,0].plot(df.loc["C"],lw=3,color = "blue")
axs[1,0].set_title("Vendas na loja C",loc = "left",fontsize = 12,fontweight=600)

axs[1,1].plot(df.loc["D"],lw=3,color = "purple")
axs[1,1].set_title("Vendas na loja D",loc = "left",fontsize = 12,fontweight=600)

for ax in axs.flat:
    ax.set_xlabel("Meses")
    ax.set_ylabel("Número de vendas")
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)

ymin = 0 
ymax = 500

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.5)
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=16,fontweight=700)
cores = ['red', 'green', 'blue', 'purple']

ymin = 0 
ymax = 500


for i, ax in enumerate(axs.flat):
    ax.plot(df.loc[df.index[i]], color=cores[i], lw=3)
    ax.set_title(f'Vendas na loja {df.index[i]}', loc='left', fontsize=12,fontweight=600)
    ax.set_xlabel('Mês', fontsize=12,fontweight=600)
    ax.set_ylabel('Número de vendas', fontsize=12,fontweight=600)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_ylim(ymin,ymax)
    
fig.savefig("vendas_4_lojas.png",transparent= False,dpi = 300,bbox_inches = "tight")
plt.show()
```
